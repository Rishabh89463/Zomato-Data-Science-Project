We would cut the Vote_Average values and make 4 categories: popular average
 below_avg not_popular to describe it more using catigorize_col() function
 provided above.

```python
def catigorize_col(df, col, labels):

    edges = [df[col].describe()['min'],
             df[col].describe()['25%'],
             df[col].describe()['50%'],
             df[col].describe()['75%'],
             df[col].describe()['max']]
    
    df[col] = pd.cut(df[col], edges , labels = labels , duplicates = 'drop')
    return df
    
```

```python
labels = ['not_popular', ' below_avg', 'average', 'popular']

catigorize_col(df, 'Vote_Average', labels)

df['Vote_Average'].unique()
```

```python
# 1. Convert the column to String first (this 'unlocks' the categorical restriction)
df['Genre'] = df['Genre'].astype(str)

# 2. Replace 'nan' or empty spots with 'Unknown'
# (Using .replace because astype(str) turns nulls into the literal word 'nan')
df['Genre'] = df['Genre'].replace('nan', 'Unknown').fillna('Unknown')

# 3. Split the strings into lists
df['Genre'] = df['Genre'].str.split(', ')

# 4. Explode the lists into separate rows
df = df.explode('Genre').reset_index(drop=True)

# 5. NOW convert to category (this creates the 'official list' based on what's there)
df['Genre'] = df['Genre'].astype('category')

# 6. Check the result
print(df['Genre'].unique())
```
